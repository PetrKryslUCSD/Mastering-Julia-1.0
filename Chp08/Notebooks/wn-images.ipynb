```julia
using Winston
wn = Winston
```

```julia
x = collect(range(0., stop=4π, length=1000));
y = collect(range(0., stop=4π, length=1000));
```

```julia
z1 = 100*(x.^0.5) .* exp.(-0.1y)';
wn.colormap("jet", 10)
wn.imagesc(z1)
```

```julia
z2 = sin.(x) .* cos.(y)';
wn.colormap("rdbu");
wn.imagesc(z2)
```
