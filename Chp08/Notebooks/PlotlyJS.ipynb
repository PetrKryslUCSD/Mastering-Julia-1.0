```julia
using PlotlyJS, DelimitedFiles, Dates
```

```julia
aaplcsv="/Users/malcolm/PacktPub/data/AAPL.csv";
(dd,dh) = readdlm(aaplcsv,','; header=true);
ddr = reverse(dd, dims=1);
```

```julia
t = Date.(ddr[end-99:end,1]);
yy = float.(ddr[end-99:end,2:5]);
lyo = Layout(;title="Apple Stock", 
                     xaxis_title="Date",
                     yaxis_title="Price");
```

```julia
plot(t,yy,lyo)
```
