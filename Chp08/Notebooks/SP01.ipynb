```julia
wF = collect(skipmissing(mlmf[mlmf.Gender .== "F", :Written]));
wM = collect(skipmissing(mlmf[mlmf.Gender .== "M", :Written]));
cF = collect(skipmissing(mlmf[mlmf.Gender .== "F", :Course]));
cM = collect(skipmissing(mlmf[mlmf.Gender .== "M", :Course]));
```
